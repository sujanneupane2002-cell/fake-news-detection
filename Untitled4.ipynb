```python


# ---------------------------------------------------------------------------
# 1. ENVIRONMENT
# ---------------------------------------------------------------------------
import os, re, string, random, time, warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, confusion_matrix, roc_curve,
                             auc, accuracy_score, precision_recall_fscore_support)

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Embedding, Bidirectional, LSTM, Dense,
                                     Dropout, GlobalMaxPooling1D)
from tensorflow.keras.callbacks import EarlyStopping

SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

for pkg in ['punkt', 'stopwords', 'wordnet', 'omw-1.4', 'punkt_tab']:
    try:
        nltk.download(pkg, quiet=True)
    except Exception:
        pass

print('=' * 60)
print(f'TensorFlow version : {tf.__version__}')
gpus = tf.config.list_physical_devices('GPU')
print(f'GPU available      : {bool(gpus)}')
print(f'Random seed        : {SEED}')
print('=' * 60)

# ---------------------------------------------------------------------------
# 2. DATASET
# ---------------------------------------------------------------------------
print('\n[Step 2] Loading ISOT dataset...')
try:
    import kagglehub
    path = kagglehub.dataset_download('clmentbisaillon/fake-and-real-news-dataset')
    real_df = pd.read_csv(f'{path}/True.csv')
    fake_df = pd.read_csv(f'{path}/Fake.csv')
except Exception as e:
    print(f'Kaggle Hub failed ({e}); falling back to local files...')
    real_df = pd.read_csv('True.csv')
    fake_df = pd.read_csv('Fake.csv')

real_df['label'] = 0  # 0 = real
fake_df['label'] = 1  # 1 = fake
df = pd.concat([real_df, fake_df], ignore_index=True)
df = df.sample(frac=1, random_state=SEED).reset_index(drop=True)
print(f'Loaded {len(df):,} articles ({len(real_df):,} real, {len(fake_df):,} fake)')

# ---------------------------------------------------------------------------
# 3. EXPLORATORY DATA ANALYSIS
# ---------------------------------------------------------------------------
print('\n[Step 3] Exploratory data analysis...')
df['text_full'] = (df['title'].fillna('') + ' ' + df['text'].fillna('')).str.strip()
df['length']    = df['text_full'].str.split().str.len()

print('Class balance:')
print(df['label'].value_counts(normalize=True).rename({0: 'real', 1: 'fake'}))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
counts = df['label'].value_counts().rename({0: 'Real', 1: 'Fake'})
axes[0].bar(counts.index, counts.values, color=['#2E75B6', '#C00060'])
axes[0].set_title('Class Distribution'); axes[0].set_ylabel('Articles')
axes[1].hist([df.loc[df.label == 0, 'length'][df.loc[df.label == 0, 'length'] < 1500],
              df.loc[df.label == 1, 'length'][df.loc[df.label == 1, 'length'] < 1500]],
             bins=60, alpha=0.7, label=['Real', 'Fake'],
             color=['#2E75B6', '#C00060'])
axes[1].set_title('Article Length (<1500 words)'); axes[1].legend()
plt.tight_layout(); plt.savefig('fig_eda.png', dpi=120); plt.show()

# ---------------------------------------------------------------------------
# 4. PREPROCESSING
# ---------------------------------------------------------------------------
print('\n[Step 4] Preprocessing...')
STOPWORDS   = set(stopwords.words('english'))
LEMMA       = WordNetLemmatizer()
URL_RE      = re.compile(r'https?://\S+|www\.\S+')
PUNCT_TABLE = str.maketrans('', '', string.punctuation)

def clean(text: str) -> str:
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = URL_RE.sub(' ', text)
    text = re.sub(r'<.*?>', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = text.translate(PUNCT_TABLE)
    tokens = word_tokenize(text)
    tokens = [LEMMA.lemmatize(t) for t in tokens
              if t not in STOPWORDS and len(t) > 2]
    return ' '.join(tokens)

from tqdm.auto import tqdm
tqdm.pandas()
df['clean'] = df['text_full'].progress_apply(clean)
print(f'Preprocessing complete. Mean clean length: '
      f'{df["clean"].str.split().str.len().mean():.0f} tokens')

# ---------------------------------------------------------------------------
# 5. TRAIN / VAL / TEST SPLIT
# ---------------------------------------------------------------------------
print('\n[Step 5] Splitting dataset...')
X = df['clean'].values
y = df['label'].values
X_tmp, X_test, y_tmp, y_test = train_test_split(
    X, y, test_size=0.10, stratify=y, random_state=SEED
)
X_train, X_val, y_train, y_val = train_test_split(
    X_tmp, y_tmp, test_size=0.111, stratify=y_tmp, random_state=SEED
)
print(f'Train: {len(X_train):,} | Val: {len(X_val):,} | Test: {len(X_test):,}')

# ---------------------------------------------------------------------------
# 6. CLASSICAL ML BASELINES (TF-IDF + 4 models)
# ---------------------------------------------------------------------------
print('\n[Step 6] Training classical models...')
vectoriser = TfidfVectorizer(ngram_range=(1, 2), max_features=50_000, min_df=5)
Xtr = vectoriser.fit_transform(X_train)
Xv  = vectoriser.transform(X_val)
Xte = vectoriser.transform(X_test)
print(f'TF-IDF shape: {Xtr.shape}')

results = {}
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, n_jobs=-1, C=1.0),
    'Multinomial NB'     : MultinomialNB(alpha=0.3),
    'Linear SVM'         : LinearSVC(C=1.0),
    'Random Forest'      : RandomForestClassifier(n_estimators=200,
                                                  n_jobs=-1, random_state=SEED),
}

print(f'{"Model":>22s} | {"Acc":>6s} | {"Prec":>6s} | {"Rec":>6s} | {"F1":>6s} | Time')
print('-' * 70)
for name, mdl in models.items():
    t0 = time.time()
    mdl.fit(Xtr, y_train)
    train_t = time.time() - t0
    pred = mdl.predict(Xte)
    acc = accuracy_score(y_test, pred)
    p, r, f1, _ = precision_recall_fscore_support(y_test, pred, average='binary')
    results[name] = {'accuracy': acc, 'precision': p, 'recall': r,
                     'f1': f1, 'train_seconds': train_t}
    print(f'{name:>22s} | {acc*100:5.2f}  | {p*100:5.2f}  | '
          f'{r*100:5.2f}  | {f1*100:5.2f}  | {train_t:5.1f}s')

# ---------------------------------------------------------------------------
# 7. BIDIRECTIONAL LSTM
# ---------------------------------------------------------------------------
print('\n[Step 7] Training BiLSTM...')
MAX_WORDS = 30_000; MAX_LEN = 300
tok = Tokenizer(num_words=MAX_WORDS, oov_token='<OOV>')
tok.fit_on_texts(X_train)

def to_seq(arr):
    return pad_sequences(tok.texts_to_sequences(arr),
                         maxlen=MAX_LEN, padding='post', truncating='post')
Str = to_seq(X_train); Sv = to_seq(X_val); Ste = to_seq(X_test)

lstm = Sequential([
    Embedding(MAX_WORDS, 100, input_length=MAX_LEN),
    Bidirectional(LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2)),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid'),
])
lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
lstm.summary()

history = lstm.fit(
    Str, y_train, validation_data=(Sv, y_val),
    epochs=10, batch_size=128,
    callbacks=[EarlyStopping(patience=2, restore_best_weights=True)],
    verbose=1,
)

lstm_proba = lstm.predict(Ste, verbose=0).ravel()
lstm_pred  = (lstm_proba > 0.5).astype(int)
lstm_acc   = accuracy_score(y_test, lstm_pred)
p, r, f1, _ = precision_recall_fscore_support(y_test, lstm_pred, average='binary')
results['BiLSTM'] = {'accuracy': lstm_acc, 'precision': p, 'recall': r, 'f1': f1}
print(f'BiLSTM: acc {lstm_acc*100:.2f} | F1 {f1*100:.2f}')

# ---------------------------------------------------------------------------
# 8. BERT-BASE FINE-TUNING
# ---------------------------------------------------------------------------
print('\n[Step 8] Fine-tuning BERT-base-uncased...')
from transformers import (BertTokenizerFast, TFBertForSequenceClassification,
                           create_optimizer)

SUBSET     = 8000          # set to None for full run
MAX_TOK    = 256
MODEL_NAME = 'bert-base-uncased'

if SUBSET is not None:
    tr_idx = np.random.RandomState(SEED).choice(len(X_train), SUBSET, replace=False)
    tr_texts, tr_labels = X_train[tr_idx], y_train[tr_idx]
else:
    tr_texts, tr_labels = X_train, y_train

btok = BertTokenizerFast.from_pretrained(MODEL_NAME)
def encode(texts):
    return btok(list(texts), padding='max_length', truncation=True,
                max_length=MAX_TOK, return_tensors='tf')
tr_enc, va_enc, te_enc = encode(tr_texts), encode(X_val), encode(X_test)

tr_ds = tf.data.Dataset.from_tensor_slices((dict(tr_enc), tr_labels)).shuffle(2048).batch(16)
va_ds = tf.data.Dataset.from_tensor_slices((dict(va_enc), y_val)).batch(32)
te_ds = tf.data.Dataset.from_tensor_slices((dict(te_enc), y_test)).batch(32)

EPOCHS = 2
steps_per_epoch = len(tr_labels) // 16
opt, _ = create_optimizer(
    init_lr=2e-5,
    num_train_steps=steps_per_epoch * EPOCHS,
    num_warmup_steps=int(0.1 * steps_per_epoch * EPOCHS),
)
bert = TFBertForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)
bert.compile(optimizer=opt,
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
bert.fit(tr_ds, validation_data=va_ds, epochs=EPOCHS)

logits     = bert.predict(te_ds, verbose=0).logits
bert_pred  = np.argmax(logits, axis=1)
bert_proba = tf.nn.softmax(logits, axis=1).numpy()[:, 1]
bert_acc   = accuracy_score(y_test, bert_pred)
p, r, f1, _ = precision_recall_fscore_support(y_test, bert_pred, average='binary')
results['BERT'] = {'accuracy': bert_acc, 'precision': p, 'recall': r, 'f1': f1}
print(f'BERT: acc {bert_acc*100:.2f} | F1 {f1*100:.2f}')

# ---------------------------------------------------------------------------
# 9. COMPARATIVE ANALYSIS
# ---------------------------------------------------------------------------
print('\n[Step 9] Generating comparison artefacts...')
summary = pd.DataFrame(results).T[['accuracy', 'precision', 'recall', 'f1']]
summary = (summary * 100).round(2).sort_values('f1', ascending=False)
print('\n' + '=' * 60)
print('FINAL RESULTS (test set, %)')
print('=' * 60)
print(summary.to_string())

# ROC curves
fig, ax = plt.subplots(figsize=(8, 6))
for name, mdl in models.items():
    if hasattr(mdl, 'predict_proba'):
        proba = mdl.predict_proba(Xte)[:, 1]
    elif hasattr(mdl, 'decision_function'):
        s = mdl.decision_function(Xte)
        proba = (s - s.min()) / (s.max() - s.min())
    else:
        continue
    fpr, tpr, _ = roc_curve(y_test, proba)
    ax.plot(fpr, tpr, label=f'{name}  AUC={auc(fpr, tpr):.3f}')
for name, proba in [('BiLSTM', lstm_proba), ('BERT', bert_proba)]:
    fpr, tpr, _ = roc_curve(y_test, proba)
    ax.plot(fpr, tpr, label=f'{name}  AUC={auc(fpr, tpr):.3f}', lw=2)
ax.plot([0, 1], [0, 1], 'k--', alpha=0.4)
ax.set_xlabel('FPR'); ax.set_ylabel('TPR'); ax.set_title('ROC Curves')
ax.legend(); plt.tight_layout(); plt.savefig('fig_roc.png', dpi=120); plt.show()

# Confusion matrix for BERT
cm = confusion_matrix(y_test, bert_pred)
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.set_title('BERT - Confusion Matrix')
plt.tight_layout(); plt.savefig('fig_confusion.png', dpi=120); plt.show()

print('\n' + classification_report(y_test, bert_pred, target_names=['Real', 'Fake']))

# ---------------------------------------------------------------------------
# 10. LIVE DEMO
# ---------------------------------------------------------------------------
print('\n[Step 10] Live demonstration...')
def predict_one(text: str):
    enc  = btok([text], padding='max_length', truncation=True,
                max_length=MAX_TOK, return_tensors='tf')
    prob = tf.nn.softmax(bert(**enc).logits, axis=1).numpy()[0]
    label = 'FAKE' if prob[1] > prob[0] else 'REAL'
    conf  = float(max(prob))
    return label, conf

demos = [
    ('REAL', 'The Bank of England held its base interest rate steady at 4.5% on Thursday, '
             'citing persistent service-sector inflation and a tight labour market as reasons '
             'for caution.'),
    ('FAKE', 'BREAKING: Scientists reveal that drinking lemon water cures all forms of cancer '
             'within 48 hours - mainstream media refuses to report it!!!'),
    ('REAL', 'Researchers at Stanford University have published a study in the journal Nature '
             'showing that a new neural-network architecture can predict protein folding with '
             'high accuracy.'),
    ('FAKE', 'SHOCK: Government CONFIRMS that 5G towers are being used to control human '
             'thoughts. Whistleblower reveals secret programme.'),
]
print('=' * 80)
for expected, text in demos:
    pred, cf = predict_one(text)
    mark = '✓' if pred == expected else '✗'
    print(f'[{mark}] Expected: {expected} | Predicted: {pred} ({cf*100:.1f}%)')
    print(f'    "{text[:90]}..."')

# ---------------------------------------------------------------------------
# 11. FINAL SYSTEM CHECK
# ---------------------------------------------------------------------------
print('\n' + '=' * 75)
print(' ' * 18 + 'FAKE NEWS DETECTION - FINAL SUMMARY')
print('=' * 75)
for name, r in sorted(results.items(), key=lambda kv: -kv[1]['f1']):
    print(f'  {name:<25s} acc {r["accuracy"]*100:5.2f}%  F1 {r["f1"]*100:5.2f}%')
print('=' * 75)
print(' ' * 22 + 'ALL SYSTEMS WORKING - DEMO COMPLETE')
print('=' * 75)
```
