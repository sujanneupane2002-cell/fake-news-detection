```python
# ---------------------------------------------------------------------------
RUN_LSTM = True
RUN_BERT = False          # Keep False on CPU. Set True only if you enable GPU.
RUN_RANDOM_FOREST = True  # Set False if you want faster run.

SEED = 42

# If you want BERT, enable GPU first, then set RUN_BERT=True.
# Runtime -> Change runtime type -> T4 GPU

# ---------------------------------------------------------------------------
# 1. ENVIRONMENT
# ---------------------------------------------------------------------------
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

if RUN_BERT:
    os.environ["TF_USE_LEGACY_KERAS"] = "1"
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "transformers", "tf-keras"])

import re, string, random, time, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_curve,
    auc,
    accuracy_score,
    precision_recall_fscore_support
)

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Input,
    Embedding,
    Bidirectional,
    LSTM,
    Dense,
    Dropout,
    GlobalMaxPooling1D
)
from tensorflow.keras.callbacks import EarlyStopping

from tqdm.auto import tqdm

random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

for pkg in ["stopwords", "wordnet", "omw-1.4"]:
    try:
        nltk.download(pkg, quiet=True)
    except Exception:
        pass

print("=" * 60)
print(f"TensorFlow version : {tf.__version__}")
gpus = tf.config.list_physical_devices("GPU")
HAS_GPU = bool(gpus)
print(f"GPU available      : {HAS_GPU}")
print(f"Random seed        : {SEED}")
print("=" * 60)

# ---------------------------------------------------------------------------
# 2. DATASET
# ---------------------------------------------------------------------------
print("\n[Step 2] Loading ISOT dataset...")

try:
    import kagglehub
except Exception:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "kagglehub"])
    import kagglehub

try:
    path = kagglehub.dataset_download("clmentbisaillon/fake-and-real-news-dataset")
    real_df = pd.read_csv(f"{path}/True.csv")
    fake_df = pd.read_csv(f"{path}/Fake.csv")
except Exception as e:
    print(f"KaggleHub failed: {e}")
    print("Trying local files: True.csv and Fake.csv")
    real_df = pd.read_csv("True.csv")
    fake_df = pd.read_csv("Fake.csv")

real_df["label"] = 0  # 0 = real
fake_df["label"] = 1  # 1 = fake

df = pd.concat([real_df, fake_df], ignore_index=True)
df = df.sample(frac=1, random_state=SEED).reset_index(drop=True)

print(f"Loaded {len(df):,} articles ({len(real_df):,} real, {len(fake_df):,} fake)")

# ---------------------------------------------------------------------------
# 3. EXPLORATORY DATA ANALYSIS
# ---------------------------------------------------------------------------
print("\n[Step 3] Exploratory data analysis...")

df["text_full"] = (
    df["title"].fillna("") + " " + df["text"].fillna("")
).str.strip()

df["length"] = df["text_full"].str.split().str.len()

print("Class balance:")
print(df["label"].value_counts(normalize=True).rename({0: "real", 1: "fake"}))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

counts = df["label"].value_counts().sort_index()
axes[0].bar(["Real", "Fake"], counts.values)
axes[0].set_title("Class Distribution")
axes[0].set_ylabel("Articles")

real_lengths = df.loc[(df.label == 0) & (df.length < 1500), "length"]
fake_lengths = df.loc[(df.label == 1) & (df.length < 1500), "length"]

axes[1].hist(
    [real_lengths, fake_lengths],
    bins=60,
    alpha=0.7,
    label=["Real", "Fake"]
)
axes[1].set_title("Article Length (<1500 words)")
axes[1].legend()

plt.tight_layout()
plt.savefig("fig_eda.png", dpi=120)
plt.show()

# ---------------------------------------------------------------------------
# 4. PREPROCESSING
# ---------------------------------------------------------------------------
print("\n[Step 4] Preprocessing...")

STOPWORDS = set(stopwords.words("english"))
LEMMA = WordNetLemmatizer()
URL_RE = re.compile(r"https?://\S+|www\.\S+")
HTML_RE = re.compile(r"<.*?>")
TOKEN_RE = re.compile(r"[a-z]{3,}")

def clean(text: str) -> str:
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = URL_RE.sub(" ", text)
    text = HTML_RE.sub(" ", text)

    tokens = TOKEN_RE.findall(text)

    tokens = [
        LEMMA.lemmatize(t)
        for t in tokens
        if t not in STOPWORDS
    ]

    return " ".join(tokens)

tqdm.pandas()
df["clean"] = df["text_full"].progress_apply(clean)

mean_len = df["clean"].str.split().str.len().mean()
print(f"Preprocessing complete. Mean clean length: {mean_len:.0f} tokens")

# ---------------------------------------------------------------------------
# 5. TRAIN / VAL / TEST SPLIT
# ---------------------------------------------------------------------------
print("\n[Step 5] Splitting dataset...")

X = df["clean"].values
y = df["label"].values

X_tmp, X_test, y_tmp, y_test = train_test_split(
    X,
    y,
    test_size=0.10,
    stratify=y,
    random_state=SEED
)

X_train, X_val, y_train, y_val = train_test_split(
    X_tmp,
    y_tmp,
    test_size=1/9,
    stratify=y_tmp,
    random_state=SEED
)

print(f"Train: {len(X_train):,} | Val: {len(X_val):,} | Test: {len(X_test):,}")

# ---------------------------------------------------------------------------
# 6. CLASSICAL ML BASELINES
# ---------------------------------------------------------------------------
print("\n[Step 6] Training classical models...")

vectoriser = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=50_000,
    min_df=5
)

Xtr = vectoriser.fit_transform(X_train)
Xv = vectoriser.transform(X_val)
Xte = vectoriser.transform(X_test)

print(f"TF-IDF shape: {Xtr.shape}")

results = {}
predictions = {}
scores_for_roc = {}
trained_models = {}

models = {
    "Logistic Regression": LogisticRegression(
        max_iter=1000,
        n_jobs=-1,
        C=1.0,
        random_state=SEED
    ),
    "Multinomial NB": MultinomialNB(alpha=0.3),
    "Linear SVM": LinearSVC(C=1.0, random_state=SEED),
}

if RUN_RANDOM_FOREST:
    models["Random Forest"] = RandomForestClassifier(
        n_estimators=100,
        n_jobs=-1,
        random_state=SEED
    )

print(f'{"Model":>22s} | {"Acc":>6s} | {"Prec":>6s} | {"Rec":>6s} | {"F1":>6s} | Time')
print("-" * 75)

for name, mdl in models.items():
    t0 = time.time()

    mdl.fit(Xtr, y_train)

    train_t = time.time() - t0
    pred = mdl.predict(Xte)

    acc = accuracy_score(y_test, pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_test,
        pred,
        average="binary",
        zero_division=0
    )

    results[name] = {
        "accuracy": acc,
        "precision": p,
        "recall": r,
        "f1": f1,
        "train_seconds": train_t
    }

    predictions[name] = pred
    trained_models[name] = mdl

    if hasattr(mdl, "predict_proba"):
        scores_for_roc[name] = mdl.predict_proba(Xte)[:, 1]
    elif hasattr(mdl, "decision_function"):
        scores_for_roc[name] = mdl.decision_function(Xte)

    print(
        f"{name:>22s} | {acc*100:5.2f}  | {p*100:5.2f}  | "
        f"{r*100:5.2f}  | {f1*100:5.2f}  | {train_t:5.1f}s"
    )

# ---------------------------------------------------------------------------
# 7. BIDIRECTIONAL LSTM
# ---------------------------------------------------------------------------
if RUN_LSTM:
    print("\n[Step 7] Training BiLSTM...")

    MAX_WORDS = 30_000
    MAX_LEN = 200 if HAS_GPU else 150
    EMBED_DIM = 100
    LSTM_UNITS = 64 if HAS_GPU else 32
    BATCH_SIZE = 128 if HAS_GPU else 256
    EPOCHS = 4 if HAS_GPU else 3

    tok = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
    tok.fit_on_texts(X_train)

    def to_seq(arr):
        return pad_sequences(
            tok.texts_to_sequences(arr),
            maxlen=MAX_LEN,
            padding="post",
            truncating="post"
        )

    Str = to_seq(X_train)
    Sv = to_seq(X_val)
    Ste = to_seq(X_test)

    if HAS_GPU:
        lstm_train_X = Str
        lstm_train_y = y_train
    else:
        # CPU-safe training subset
        LSTM_TRAIN_LIMIT = min(12000, len(Str))
        sub_idx = np.random.RandomState(SEED).choice(
            len(Str),
            LSTM_TRAIN_LIMIT,
            replace=False
        )
        lstm_train_X = Str[sub_idx]
        lstm_train_y = y_train[sub_idx]
        print(f"CPU mode: training BiLSTM on {LSTM_TRAIN_LIMIT:,} samples for speed.")

    lstm = Sequential([
        Input(shape=(MAX_LEN,)),
        Embedding(MAX_WORDS, EMBED_DIM),
        Bidirectional(
            LSTM(
                LSTM_UNITS,
                return_sequences=True,
                dropout=0.3,
                recurrent_dropout=0.0
            )
        ),
        GlobalMaxPooling1D(),
        Dense(64, activation="relu"),
        Dropout(0.5),
        Dense(1, activation="sigmoid"),
    ])

    lstm.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    lstm.summary()

    t0 = time.time()

    history = lstm.fit(
        lstm_train_X,
        lstm_train_y,
        validation_data=(Sv, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[
            EarlyStopping(
                monitor="val_loss",
                patience=1,
                restore_best_weights=True
            )
        ],
        verbose=1
    )

    lstm_train_t = time.time() - t0

    lstm_proba = lstm.predict(Ste, verbose=0).ravel()
    lstm_pred = (lstm_proba > 0.5).astype(int)

    lstm_acc = accuracy_score(y_test, lstm_pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_test,
        lstm_pred,
        average="binary",
        zero_division=0
    )

    results["BiLSTM"] = {
        "accuracy": lstm_acc,
        "precision": p,
        "recall": r,
        "f1": f1,
        "train_seconds": lstm_train_t
    }

    predictions["BiLSTM"] = lstm_pred
    scores_for_roc["BiLSTM"] = lstm_proba

    print(f"BiLSTM: acc {lstm_acc*100:.2f}% | F1 {f1*100:.2f}% | Time {lstm_train_t:.1f}s")

else:
    print("\n[Step 7] BiLSTM skipped.")

# ---------------------------------------------------------------------------
# 8. OPTIONAL BERT
# ---------------------------------------------------------------------------
if RUN_BERT:
    print("\n[Step 8] Fine-tuning BERT-base-uncased...")

    from transformers import (
        BertTokenizerFast,
        TFBertForSequenceClassification,
        create_optimizer
    )

    SUBSET = 8000 if HAS_GPU else 1500
    MAX_TOK = 256 if HAS_GPU else 128
    MODEL_NAME = "bert-base-uncased"
    BERT_EPOCHS = 2 if HAS_GPU else 1
    BERT_BATCH = 16 if HAS_GPU else 8

    tr_idx = np.random.RandomState(SEED).choice(
        len(X_train),
        min(SUBSET, len(X_train)),
        replace=False
    )

    tr_texts = X_train[tr_idx]
    tr_labels = y_train[tr_idx]

    btok = BertTokenizerFast.from_pretrained(MODEL_NAME)

    def encode(texts):
        return btok(
            list(texts),
            padding="max_length",
            truncation=True,
            max_length=MAX_TOK,
            return_tensors="tf"
        )

    tr_enc = encode(tr_texts)
    va_enc = encode(X_val)
    te_enc = encode(X_test)

    tr_ds = (
        tf.data.Dataset
        .from_tensor_slices((dict(tr_enc), tr_labels))
        .shuffle(2048, seed=SEED)
        .batch(BERT_BATCH)
    )

    va_ds = (
        tf.data.Dataset
        .from_tensor_slices((dict(va_enc), y_val))
        .batch(32)
    )

    te_ds = (
        tf.data.Dataset
        .from_tensor_slices((dict(te_enc), y_test))
        .batch(32)
    )

    steps_per_epoch = max(1, len(tr_labels) // BERT_BATCH)

    opt, _ = create_optimizer(
        init_lr=2e-5,
        num_train_steps=steps_per_epoch * BERT_EPOCHS,
        num_warmup_steps=int(0.1 * steps_per_epoch * BERT_EPOCHS)
    )

    bert = TFBertForSequenceClassification.from_pretrained(
        MODEL_NAME,
        num_labels=2
    )

    bert.compile(
        optimizer=opt,
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        metrics=["accuracy"]
    )

    t0 = time.time()

    bert.fit(
        tr_ds,
        validation_data=va_ds,
        epochs=BERT_EPOCHS,
        verbose=1
    )

    bert_train_t = time.time() - t0

    logits = bert.predict(te_ds, verbose=0).logits
    bert_pred = np.argmax(logits, axis=1)
    bert_proba = tf.nn.softmax(logits, axis=1).numpy()[:, 1]

    bert_acc = accuracy_score(y_test, bert_pred)
    p, r, f1, _ = precision_recall_fscore_support(
        y_test,
        bert_pred,
        average="binary",
        zero_division=0
    )

    results["BERT"] = {
        "accuracy": bert_acc,
        "precision": p,
        "recall": r,
        "f1": f1,
        "train_seconds": bert_train_t
    }

    predictions["BERT"] = bert_pred
    scores_for_roc["BERT"] = bert_proba

    print(f"BERT: acc {bert_acc*100:.2f}% | F1 {f1*100:.2f}% | Time {bert_train_t:.1f}s")

else:
    print("\n[Step 8] BERT skipped. Set RUN_BERT=True only with GPU.")

# ---------------------------------------------------------------------------
# 9. COMPARATIVE ANALYSIS
# ---------------------------------------------------------------------------
print("\n[Step 9] Generating comparison artefacts...")

summary = pd.DataFrame(results).T

display_cols = ["accuracy", "precision", "recall", "f1"]
summary_display = (summary[display_cols] * 100).round(2)
summary_display = summary_display.sort_values("f1", ascending=False)

print("\n" + "=" * 60)
print("FINAL RESULTS - TEST SET (%)")
print("=" * 60)
print(summary_display.to_string())

summary_display.to_csv("model_results_summary.csv")

# ROC curves
fig, ax = plt.subplots(figsize=(8, 6))

for name, score in scores_for_roc.items():
    fpr, tpr, _ = roc_curve(y_test, score)
    model_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f"{name} AUC={model_auc:.3f}")

ax.plot([0, 1], [0, 1], "k--", alpha=0.4)
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.set_title("ROC Curves")
ax.legend()
plt.tight_layout()
plt.savefig("fig_roc.png", dpi=120)
plt.show()

# Best model confusion matrix
best_model_name = summary_display.index[0]
best_pred = predictions[best_model_name]

cm = confusion_matrix(y_test, best_pred)

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Real", "Fake"],
    yticklabels=["Real", "Fake"]
)
ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
ax.set_title(f"{best_model_name} - Confusion Matrix")
plt.tight_layout()
plt.savefig("fig_confusion.png", dpi=120)
plt.show()

print(f"\nClassification Report - {best_model_name}")
print(classification_report(
    y_test,
    best_pred,
    target_names=["Real", "Fake"],
    zero_division=0
))

# ---------------------------------------------------------------------------
# 10. LIVE DEMO
# ---------------------------------------------------------------------------
print("\n[Step 10] Live demonstration...")

def sigmoid(x):
    x = np.clip(x, -50, 50)
    return 1 / (1 + np.exp(-x))

def predict_one(text: str):
    cleaned = clean(text)

    # Use the best model automatically
    if best_model_name in trained_models:
        mdl = trained_models[best_model_name]
        vec = vectoriser.transform([cleaned])

        pred = int(mdl.predict(vec)[0])

        if hasattr(mdl, "predict_proba"):
            fake_prob = float(mdl.predict_proba(vec)[0, 1])
        elif hasattr(mdl, "decision_function"):
            score = float(mdl.decision_function(vec)[0])
            fake_prob = float(sigmoid(score))
        else:
            fake_prob = float(pred)

        label = "FAKE" if pred == 1 else "REAL"
        conf = fake_prob if pred == 1 else 1 - fake_prob
        return label, conf

    elif best_model_name == "BiLSTM":
        seq = pad_sequences(
            tok.texts_to_sequences([cleaned]),
            maxlen=MAX_LEN,
            padding="post",
            truncating="post"
        )
        fake_prob = float(lstm.predict(seq, verbose=0).ravel()[0])
        label = "FAKE" if fake_prob > 0.5 else "REAL"
        conf = fake_prob if label == "FAKE" else 1 - fake_prob
        return label, conf

    elif best_model_name == "BERT":
        enc = btok(
            [text],
            padding="max_length",
            truncation=True,
            max_length=MAX_TOK,
            return_tensors="tf"
        )
        prob = tf.nn.softmax(bert(**enc).logits, axis=1).numpy()[0]
        label = "FAKE" if prob[1] > prob[0] else "REAL"
        conf = float(max(prob))
        return label, conf

    else:
        raise ValueError("No valid model found for prediction.")

demos = [
    (
        "REAL",
        "The Bank of England held its base interest rate steady on Thursday, citing persistent inflation and labour market conditions."
    ),
    (
        "FAKE",
        "BREAKING: Scientists reveal that drinking lemon water cures all forms of cancer within 48 hours."
    ),
    (
        "REAL",
        "Researchers at Stanford University published a study showing that a new neural network architecture improved protein prediction accuracy."
    ),
    (
        "FAKE",
        "SHOCK: Government confirms that 5G towers are being used to control human thoughts, according to secret sources."
    ),
]

print("=" * 80)
print(f"Using best model: {best_model_name}")
print("=" * 80)

for expected, text in demos:
    pred, cf = predict_one(text)
    mark = "✓" if pred == expected else "✗"

    print(f"[{mark}] Expected: {expected} | Predicted: {pred} ({cf*100:.1f}%)")
    print(f'    "{text[:100]}..."')

# ---------------------------------------------------------------------------
# 11. FINAL SYSTEM CHECK
# ---------------------------------------------------------------------------
print("\n" + "=" * 75)
print(" " * 18 + "FAKE NEWS DETECTION - FINAL SUMMARY")
print("=" * 75)

for name, r in sorted(results.items(), key=lambda kv: -kv[1]["f1"]):
    train_time = r.get("train_seconds", 0)
    print(
        f'  {name:<25s} acc {r["accuracy"]*100:5.2f}%  '
        f'F1 {r["f1"]*100:5.2f}%  time {train_time:7.1f}s'
    )

print("=" * 75)
print(" " * 22 + "ALL SYSTEMS WORKING - DEMO COMPLETE")
print("=" * 75)
```
